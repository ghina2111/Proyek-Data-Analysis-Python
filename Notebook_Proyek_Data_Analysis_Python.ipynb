```python
air1_df_numeric = air1_df.select_dtypes(include=['float64', 'int64'])

Q1 = air1_df_numeric.quantile(0.25)
Q3 = air1_df_numeric.quantile(0.75)
IQR = Q3 - Q1

outliers = ((air1_df_numeric < (Q1 - 1.5 * IQR)) | (air1_df_numeric > (Q3 + 1.5 * IQR)))

outliers_data = air1_df_numeric[outliers.any(axis=1)]

outliers_count_per_column = outliers.sum()

total_outliers = outliers_count_per_column.sum()

if total_outliers > 0:
    print(f"Outliers ditemukan di {len(outliers_data)} baris.")
    print("Jumlah outlier per kolom:")
    print(outliers_count_per_column)
    print(f"\nTotal outlier di seluruh kolom: {total_outliers}")
else:
    print("Tidak ada outlier yang ditemukan.")
```

SOLVING OUTLIER AND RESULT

```python
def replace_outliers(df):
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        maximum = Q3 + (1.5 * IQR)
        minimum = Q1 - (1.5 * IQR)

        df[column] = df[column].mask(df[column] > maximum, maximum)
        df[column] = df[column].mask(df[column] < minimum, minimum)

replace_outliers(air1_df)
replace_outliers(air2_df)
replace_outliers(air3_df)
replace_outliers(air4_df)
replace_outliers(air5_df)
replace_outliers(air6_df)
replace_outliers(air7_df)
replace_outliers(air8_df)
replace_outliers(air9_df)
replace_outliers(air10_df)
replace_outliers(air11_df)
replace_outliers(air12_df)

print(f"DataFrame air1_df:\n{air1_df.head()}\n")
print(f"DataFrame air2_df:\n{air2_df.head()}\n")
print(f"DataFrame air3_df:\n{air3_df.head()}\n")
print(f"DataFrame air4_df:\n{air4_df.head()}\n")
print(f"DataFrame air5_df:\n{air5_df.head()}\n")
print(f"DataFrame air6_df:\n{air6_df.head()}\n")
print(f"DataFrame air7_df:\n{air7_df.head()}\n")
print(f"DataFrame air8_df:\n{air8_df.head()}\n")
print(f"DataFrame air9_df:\n{air9_df.head()}\n")
print(f"DataFrame air10_df:\n{air10_df.head()}\n")
print(f"DataFrame air11_df:\n{air11_df.head()}\n")
print(f"DataFrame air12_df:\n{air12_df.head()}\n")
```

Menginisialisasi pollutans dan membentuk DataFrame baru yang berisi rata-rata nilai polutan untuk setiap stasiun pengukuran pada tahun 2017, dan menyimpannya dalam variabel city_pollutant_avg_2017

```python
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
city_pollutant_avg_2017 = df_all_2017.groupby('station')[pollutants].mean().reset_index()
```

Menghitung total polusi partikel (PM2.5 dan PM10) untuk setiap stasiun, kemudian mengurutkan hasilnya untuk mengidentifikasi kota-kota dengan tingkat polusi tertinggi pada tahun 2017

```python
city_pollutant_avg_2017['Total_PM'] = city_pollutant_avg_2017['PM2.5'] + city_pollutant_avg_2017['PM10']
top_polluted_cities_2017 = city_pollutant_avg_2017.sort_values(by='Total_PM', ascending=False)
```

```python
print("Top 5 kota dengan kualitas udara terburuk di tahun 2017 (berdasarkan PM2.5 + PM10):")
print(top_polluted_cities_2017[['station', 'PM2.5', 'PM10', 'Total_PM']].head(5))
```

```python
all_cities_df_2013_2017 = df_all[df_all['year'].between(2013, 2017)]
```

inisialisasi pollutans

```python
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
```

Menghitung rata-rata nilai polutan untuk setiap kombinasi tahun dan bulan dari data yang telah difilter untuk tahun 2013-2017.

```python
monthly_avg = all_cities_df_2013_2017.groupby(['year', 'month'])[pollutants].mean().reset_index()
```

Menambahkan kolom baru yang merepresentasikan total polutan di setiap kombinasi tahun dan bulan

```python
monthly_avg['Total_Pollutants'] = monthly_avg['PM2.5'] + monthly_avg['PM10'] + monthly_avg['SO2'] + monthly_avg['NO2'] + monthly_avg['CO'] + monthly_avg['O3']
```

```python
for year in range(2013, 2017+1):
    print(f"Bulan dengan kualitas udara terburuk di tahun {year} (berdasarkan Total Polutan):")
    print(monthly_avg[monthly_avg['year'] == year][['month', 'Total_Pollutants']].sort_values(by='Total_Pollutants', ascending=False).head())
```

**Insight:**
- Terdapat 5 kota dengan kualitas udara terburuk sepanjang tahun 2017.
- Ada periode(waktu/bulan)tertentu di setiap tahunnya (2013-2017) yang mengalami kualitas udara buruk
